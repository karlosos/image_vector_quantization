```python
import pandas as pd
import numpy as np

k256_df = pd.read_csv('compression_256_4.csv', index_col='Unnamed: 0')
k512_df = pd.read_csv('compression_512_4.csv', index_col='Unnamed: 0')
k1024_df = pd.read_csv('compression_1024_4.csv', index_col='Unnamed: 0')
```

```python
def reduce_df(df, name):    
    df["size_comp"] = df["size_golomb"] + df["size_quant"]
    df["cr"] = df["size_bmp"]/df["size_comp"]
    res = df.rename(columns = {'size_comp': f'size_comp_{name}', 'cr': f'cr_{name}', 'psnr': f'psnr_{name}'})[[f"cr_{name}", f"psnr_{name}"]]
    return res
```

```python
df_0 = reduce_df(k256_df, '256')
df_1 = reduce_df(k512_df, '512')
df_2 = reduce_df(k1024_df, '1024')
```

```python
df = pd.concat([k256_df[["name", "size"]], df_0, df_1, df_2], axis=1)
```
