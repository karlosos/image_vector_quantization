# Porównanie naszego rozwiązania z JPG

```python
import pandas as pd
```

```python
df1 = pd.read_csv("jpg_compression_256_4.csv", index_col='Unnamed: 0')
df2 = pd.read_csv("jpg_compression_512_4.csv", index_col='Unnamed: 0')
df3 = pd.read_csv("jpg_compression_1024_4.csv", index_col='Unnamed: 0')

df = pd.concat([df1, df2.loc[:, "size_our":"psnr_jpg"], df3.loc[:, "size_our":"psnr_jpg"]], axis=1)
df.loc[:, ["size_our", "size_jpg"]] = df.loc[:, ["size_our", "size_jpg"]]/8000
df
```

```python
print(df.round(4).to_latex(index=False))
```

```python
df_agg = df.groupby("image_size").mean().iloc[:, [2, 3, 6, 7, 10, 11]]
df_agg
```

```python
print(df_agg.to_latex())
```
